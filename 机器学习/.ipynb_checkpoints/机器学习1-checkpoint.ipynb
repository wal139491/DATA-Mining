## KNN(K-近邻算法)

K近邻算法就是测量不同特征值之间距离方法进行分类

>工作原理：
- 存在一个样本数据集合(训练数据集)，样本数据集中每个数据都有标签，标签（样本数据集中的每个数据与所属分类的对应关系）
- 输入没有标签的数据，将新数据与样本集中数据对应的特征进行比较
- 算法提取样本集中数据最相似（最近邻）数据的分类标签
   一般只取k个最相似的数据（k是 $\lt$ 20 的整数 ）
- 最后取出k个相似数据中出现次数最多的分类，就是新数据的分类

#### KNN算法的伪代码

对未知类别属性的数据集中的每个点依次执行以下的操作：

1. 计算已知类别数据集中的点与当前点之间的距离
2. 按照距离大小递增进行排序
3. 选取与当前点距离最小的k个点
4. 确定前k个点**类别**的出现频率
5. 返回出现频率最高的类别作为当前点的预测分类

````python
from numpy import *
import operator
from os import listdir

````

1. 计算数据集中特征值(*一个向量点包含多个特征值*)之间距离
计算两个向量点$xA$,$xB$之间距离：
</br>
$$d = \sqrt {{(xA_0 - xB_0)^2} + {(xA_1 - xB_1)^2}}$$

如果数据集中存在4个特征值，那么点(1,0,0,1)与(7,6,9,4)之间的距离是：
$$\sqrt {(7 - 1)^2 + (6 - 0)^2 + (9 - 0)^2 + (4 - 1)^2}$$

2. 对计算的距离进行排序，确定前k（k是正整数）个最小距离元素的主要分类
3. 将classCount字典分解为元组列表，然后使用operator模块的itemgetter方法，
   按照第二个元素次序对元组排序（逆序），此时返回最大频率的分类


````python
'''
classify0()函数有4个输入参数
inX:  用于分类的输入向量
dataSet: 训练样本集
labels: 标签向量
k: 最近邻居的数目
'''

def classify0(inX, dataSet, labels, k):
    dataSetSize = dataSet.shape[0]
    # 计算特征值(向量点)距离
    diffMat = tile(inX, (dataSetSize,1)) - dataSet #输入向量-数据集点
    sqDiffMat = diffMat**2
    sqDistances = sqDiffMat.sum(axis=1)
    distances = sqDistances**0.5
    sortDistDistances = distances.argsort()
    classCount = {}  #保存分类的数量
    # 选择k个距离最小的点
    for i in range(k):
        voteIlabel = labels[sortDistDistances[i]] #找出k个近邻的标签保存在列表中
        classCount[voteIlabel] = classCount.get(voteIlabel,0) + 1
        # get返回classCount指定key的值，没有默认为0
        # 如果k近邻中存在这个分类，就+1
    sortedClassCount = sorted(classCount.iteritems(), key=operator.itemgetter(1), reverse=True)#按第二个元素逆序
    return sortedClassCount[0][0]
    
    
````

实现KNN算法步骤：

4. 训练算法：此步骤不适用于kNN算法
5. 测试算法：使用用户提供的部分数据作为测试样本
   测试样本与非测试样本的区别在于，测试样本是已经完成分类的数据，如果预测分类与实际类别不同，则标记为一个错误
6. 使用算法：产生简单的命令行程序，用户输入一些特征数据，判断对方是否为自己喜欢的类型。



````python
'''
为了将.txt数据转换成分类器能够处理的格式
创建file2matirx()函数处理输入格式问题
输入文件名字符串，返回*训练样本矩阵*和*类标签向量*
'''
def file2matrix(filename):
    fr = open(filename)
    arrayOfLines = fr.readlines()
    numberOfLines = len(arrayOfLines) # 得到文件行数
    returnMat = zeros((numberOfLines,3)) # 返回创建的Numpy矩阵,3是设置的另一维度值
    classLabelVector = [] #类标签向量
    index = 0
    # 解析文件数据到列表
    for line in arrayOfLines:
        line = line.strip()
        listFromLine = line.split('\t')
        # 每行的元素分割放在一个列表中
        returnMat[index,:]= listFromLine[:3]
        # 选取前3个元素存储在特征矩阵中
        classLabelVector.append(int(listFromLine[-1]))
        # 每次取最后一列存储在classLabelvec列表中
        index += 1
    return returnMat,classLabelVector
    
````
