## MovieLens 1M数据集

```python
import pandas as pd

unames = ['user_id', 'gender', 'age' , 'occupation', 'zip']
users = pd.read_table(r'F:\数据挖掘\pydata-book\ch02\movielens\users.dat', sep='::', header=None, names=unames)

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table(r'F:\数据挖掘\pydata-book\ch02\movielens\ratings.dat', sep='::', header=None, names=rnames)

mnames  = ['movie_id', 'title', 'geners']
movies = pd.read_table(r'F:\数据挖掘\pydata-book\ch02\movielens\movies.dat', sep='::', header=None, names=mnames)
```

```python
data = pd.merge(pd.merge(ratings, users), movies)
# merge根据列名重叠情况，推断那些列是合并的
```

```python
data
```

```python
data.ix[0]
```

```python
mean_ratings = data.pivot_table('rating', index='title', columns  = 'gender', aggfunc='mean')
# 产生一个DataFrame,行为电影名称，列为性别，内容是电影平均得分
# female male
```

```python
mean_ratings[:5]
```

```python
# 先过滤掉评论不足250条的电影
# 按照电影名称进行分组，在用size对分组进行计数

ratings_by_title = data.groupby('title').size()
# 利用size得到一个含有各电影分组大小的Series对象
ratings_by_title[:10]
```

```python
active_titles = ratings_by_title.index[ratings_by_title >= 250]
```

```python
active_titles
# 得到评论数大于250的索引，这些索引的名称是电影名，因为上面是按电影名进行分组的
```

```python
# 根据上面索引，从mean_rating选取需要的行
mean_ratings = mean_ratings.ix[active_titles]
```

```python
mean_ratings.head()
# 获取评论大于250，然后根据性别计算每个电影的平均得分
```

```python
# 为了了解女性观众最喜欢的电影，按降序排列
top_female_movies = mean_ratings.sort_values(by='F', ascending=False)
top_female_movies[:10]
```

```python
top_male_movies = mean_ratings.sort_values(by = 'M', ascending=False)
top_male_movies[:10]
```

### 计算平均分歧
想找出男女分歧最大的电影，
1. mean_ratings增加一个男女电影评分之差的列
2. 对这个列进行排序

```python
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
sort_by_diff = mean_ratings.sort_values(by = 'diff')
```

```python
sort_by_diff[:15]
```

```python
# 对sort_by_diff取反序就得到男性最喜欢的电影
sort_by_diff[::-1][:15]
# [::-1] 因为有三列数据，就对最后一列进行反序，在切片得到前15行数据
```

```python
# 根据电影名称分组的得分数据的标准差
rating_std_by_title = data.groupby('title')['rating'].std()
print(type(rating_std_by_title))
```

```python
# 根据active_title进行过滤
rating_std_by_title = rating_std_by_title.ix[active_titles]
```

```python
# 根据值对Series进行降序排列
rating_std_by_title.sort_values(ascending=False)[:10]
```
